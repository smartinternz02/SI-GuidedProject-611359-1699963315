# Data preprocessing

```python
data.drop(['Timestamp','comments'] ,axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Country'].value_counts().plot(kind='bar',figsize=(10,8))
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Counts of Each Country')
plt.show()
```

```python
# Assuming data is your DataFrame or dataset
data['state'].value_counts().plot(kind='bar', figsize=(10, 8))
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Counts of Each State')
plt.show()
```

```python
data.drop(['Country','state'],axis=1,inplace=True)
```

```python
data.head()
```

# Checking null values

```python
data.isnull().sum()
```

```python
data['self_employed'].mode()
```

```python
data['work_interfere'].mode()
```

```python
data['work_interfere'].fillna('N/A', inplace=True)
data['self_employed'].fillna('No', inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Age'].value_counts().plot(kind='bar',figsize=(10,8))
```

```python
#so only keeping the rows with the age between 18 and 60
data.drop(data[(data['Age']>60)| (data['Age']<18)].index,inplace=True)
```

```python
data['Gender'].value_counts().plot(kind='bar',figsize=(10,8))
```

```python
data['Gender'].value_counts().plot(kind='bar',figsize=(10,8))
```

```python
data['Gender'].values
```
